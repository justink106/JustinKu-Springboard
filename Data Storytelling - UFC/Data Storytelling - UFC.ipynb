```python
# Importing modules

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing my dataset

ufc = pd.read_csv('clean_ufc_all_fights.csv')
```

```python
# Making a fight ID column from the index

ufc = ufc.reset_index().rename(columns={'index':'fight_id'})
```

```python
# Creating a list of features that will be compared

features = [
    'Weight', 'Reach', 'Age', 'Height_in',
    'Stance', 'SLpM', 'StrAcc', 'StrDef',
    'TDAvg', 'TDAcc', 'TDDef', 'SubAvg'
]
```

```python
# Creating a function to pull out each fighter into their own row

def make_long(ufc, n):
    pref = f'fighter{n}_'
    cols = [pref + f for f in features]
    sub = ufc[['fight_id', 'fight_outcome'] + cols].copy()

    sub.columns = ['fight_id', 'fight_outcome'] + features
    sub['fighter_num'] = n
    return sub
```

```python
# Creating the long dataframe

ufc1 = make_long(ufc, 1)
ufc2 = make_long(ufc, 2)
long_ufc = pd.concat([ufc1, ufc2], ignore_index=True)
```

```python
# Turning fight_outcome into a numeric win flag

long_ufc['win'] = (long_ufc['fighter_num'] == long_ufc['fight_outcome'].str[-1].astype(int))
```

### Looking at the win rate by stance

```python
plt.figure(figsize=(6,4))
sns.barplot(data=long_ufc, x='Stance', y='win', errorbar=None)
plt.title('Win Rate by Stance')
plt.ylabel('Win Rate')
plt.ylim(0,1)
plt.show()
```

### Looking at the body-parameters

```python
body_cols = ['Weight','Reach','Height_in','Age']
fig, axes = plt.subplots(1, len(body_cols), figsize=(len(body_cols)*4, 4))

for ax, col in zip(axes, body_cols):
    sns.boxplot(data=long_ufc, x='win', y=col, ax=ax)
    ax.set_xlabel('')              # hide the “win” label
    ax.set_title(col)

fig.suptitle('Body Parameters: Winners (1) vs. Losers (0)', y=1.05)
plt.tight_layout()
plt.show()
```

### Looking at skill-metrics

```python
skill_cols = [
    'SLpM','StrAcc', 'StrDef',
    'TDAvg','TDAcc','TDDef','SubAvg'
]

fig, axes = plt.subplots(4, 2, figsize=(10, 12))
axes = axes.flatten()

for ax, col in zip(axes, skill_cols):
    sns.boxplot(data=long_ufc, x='win', y=col, ax=ax)
    ax.set_xlabel('')
    ax.set_title(col.replace('_',' '))

fig.suptitle('Skill Metrics: Winners (1) vs. Losers (0)', y=1.02)
plt.tight_layout()
plt.show()
```
